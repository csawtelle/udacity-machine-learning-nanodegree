Question 1
Given two strings s and t, determine whether some anagram of t is a substring of s. For example: if s = "udacity" and t = "ad", then the function returns True. Your function definition should look like: question1(s, t) and return a boolean True or False.

```python
# Function call takes in two strings
def question1(s,t):
    # Those strings are converted to sets
    set1 = set(s)
    set2 = set(t)

    # Subtract set1 from set2
    # If set2 has elements that set1 does not, it won't be empty
    if not set2 - set1: 
        return True
    else:
        return False

# Test 1
s = "udacity"
t = "ad"   
result = question1(s,t)
print(result)

# Test 2
s = "udacity"
t = "adx"   
result = question1(s,t)
print(result)
```
